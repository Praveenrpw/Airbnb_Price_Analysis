```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

# Data Cleaning

```python

data.isnull().sum()
```

```python
# use to remove all null values rows

#data.dropna(inplace = True)
#data
```

```python
# Replace null values in 'price' with the mean of the 'price' column

data.fillna({'price': data['price'].mean()}, inplace=True)
```

```python
# verify is the null values present in the price column

data.isnull().sum()
```

```python
# remove the rows of other null value present column

data.dropna(inplace = True)
```

```python
# verifying anymore null values, all null values cleared

data.isnull().sum()
```

```python
# checking the number of rows and columns after removing null rows

data.shape
```

```python
# next checking with any duplicates rows
# there are 12 duplicated records

data.duplicated().sum()
```

```python
# displaying the entire duplicates, repeated records

data[data.duplicated()]
```

```python
# removing all duplicated rows

data.drop_duplicates(inplace = True)
```

```python
# verifying all duplicated rows are removed

data.duplicated().sum()
```

### Type Casting

```python
data.dtypes
```

```python
# type casting 'id', 'host_id' to object

data[['id', 'host_id']] = data[['id', 'host_id']].astype(object)
```

# Data Analysis EDA

### Univariate Analysis

```python
# Accessing the price column

data['price']
```

```python
# Finding the distribution of prices, This indicates the presence of extreme outliers

plt.figure(figsize = (8, 5))
sns.histplot(data = data, x = 'price', bins = 100)
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualizing price distribution to identify outliers using a box plot

sns.boxplot(data = data, x = 'price')

# Most prices are within the range of 0 to 20,000
# Outliers are present beyond this range
```

```python
# Filter prices less than or equal to 1500 to remove outliers

df = data['price'] <= 1500
df
```

```python
# Filter prices less than 1500 to remove outliers

df = data[data['price'] < 1500]
sns.boxplot(data = df, x = 'price')

# Dots are uniformly distributed, indicating fewer outliers
```

```python
# Create a new dataframe as df with filtered data
# Plot the price distribution

plt.figure(figsize = (8, 5))
sns.histplot(data = df, x = 'price', bins = 100)
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# Price distribution with KDE
# kde=True adds the density curve

plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='price', bins=100, kde=True)  
plt.title('Price Distribution with KDE')
plt.ylabel('Frequency')
plt.show()
```

```python
df.columns
```

```python
df.dtypes
```

```python
# Plotting histograms for 'price' and 'availability_365' columns

columns = ['price', 'availability_365']

for column in columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x=column)
    plt.title(f'{column} Distribution')
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Grouping the data by 'neighbourhood_group' and calculating the avg price for each group

df.groupby(by = 'neighbourhood_group')['price'].mean()
```

# Feature Engineering


```python
# Creating a new column 'price_per_bed' by dividing 'price' by 'beds'

df['price_per_bed'] = df['price'] / df['beds']
df.head(3)
```

```python
# price per bed

df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
# average price per bed

df.groupby(by='neighbourhood_group')['price_per_bed'].mean()
```

### Bivariate Analysis

```python

df.columns
```

```python
# relationship between price dependency on neighbourhood

sns.barplot(data = df, x = 'neighbourhood_group', y = 'price', hue = 'room_type')
```

```python
# relationship between number of reviews and price

plt.figure(figsize = (10, 6))
plt.title('Locality and Review Dependency')
sns.scatterplot(data = df, x = 'number_of_reviews', y = 'price', hue = 'neighbourhood_group')
plt.show()
```

```python

df.columns
```

```python
# Visualizing the relationships between selected columns and room type using pairplot

sns.pairplot(data = df, vars = ['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue = 'room_type')         
```

```python
# Plotting the geographical distribution of Airbnb listings based on longitude and latitude

plt.figure(figsize = (10, 7))
sns.scatterplot(data = df, x = 'longitude', y = 'latitude', hue = 'room_type')
plt.title("Geographical Distribution of AirBnb Listing")
plt.show()
```

```python
# Heatmap showing the correlation between numerical variables

corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds', 'price_per_bed']].corr()
corr
```

```python
# Plotting a heatmap to visualize the correlation matrix with annotations
plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True, linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
